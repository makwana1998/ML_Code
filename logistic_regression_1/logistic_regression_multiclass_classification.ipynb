# Logistic Regression Multiclass Classification


```python
# importing datasets from sklearn

from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
all_digits = load_digits()
```

```python
# display digits

plt.gray()
for i in range(6):
    plt.matshow(all_digits.images[i])
```

```python
print(dir(all_digits))
```

```python
all_digits.data[0]
```

```python
# create and test logistic regression model

# importing logistic regression class

from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(all_digits.data, all_digits.target, test_size=0.2)
```

```python
log_model.fit(X_train, Y_train)
```

```python
# Measuring accuracy of our logistic model

log_model.score(X_test, Y_test)
```

```python
# predict data

log_model.predict(all_digits.data[0:8])
```

```python
# Confusion Matrix

Y_predicted = log_model.predict(X_test)
```

```python
Y_predicted
```

```python
# imporint metrics 

from sklearn.metrics import confusion_matrix
```

```python
con_mat = confusion_matrix(Y_test, Y_predicted)
```

```python
con_mat
```

```python
plt.figure(figsize = (10, 8))
```
