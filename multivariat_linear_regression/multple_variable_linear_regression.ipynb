### Given home prices find out price of a home

```python
# importing libraries

import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("house_prices.csv")
```

```python
df
```

```python
# data preprocessing: Fill NaN values with median of a column

df.bhk
```

```python
# get the median values of bhk column

df.bhk.median()
```

```python
# fillna(value)

df.bhk = df.bhk.fillna(df.bhk.median())
```

```python
df.bhk
```

```python
df
```

```python
# creating model

lin_reg = linear_model.LinearRegression()
```

```python
new_df = df.drop('price', axis='columns')
```

```python
new_df
```

```python
lin_reg.fit(new_df, df.price)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
# get price  of house with 3600 sqft area, 3 bhk, 20 year old


lin_reg.predict([[3600, 3, 20]])
```
