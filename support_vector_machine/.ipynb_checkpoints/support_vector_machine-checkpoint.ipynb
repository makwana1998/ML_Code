# Support Vector Machine

> ## A support vector machine (SVM) is a type of deep learning algorithm that performs supervised learning for classification or regression of data groups. In AI and machine learning, supervised learning systems provide both input and desired output data, which are labeled for classification.

```python
# train a model using SVM

# importing libraries from sklearn

from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target', 'flower_name'], axis = 'columns')
```

```python
Y = df.target
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train
```

```python
Y_train
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# importing SVC from sklearn

from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
model.predict([[3.0,0.2,4.6,1.2]])
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```

```python
# Tuning parameters

# Regularization (C)

model_C = SVC(C=1)
model_C.fit(X_train, Y_train)
```

```python
model_C.score(X_test, Y_test)
```

```python
model_C = SVC(C=8)
model_C.fit(X_train, Y_train)
```

```python
model_C.score(X_test, Y_test)
```

```python
# Gamma

model_g = SVC(gamma=8)
model_g.fit(X_train, Y_train)
```

```python
model_g.score(X_test, Y_test)
```

```python
# Kernal

model_linear_kernel = SVC(kernel='linear')
model_linear_kernel.fit(X_train, Y_train)
```

```python
model_linear_kernel.score(X_test, Y_test)
```
