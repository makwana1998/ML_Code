```python
# Decision Tree

# importing pandas

import pandas as pd
```

```python
# importing dataset

df = pd.read_csv('company.csv')
```

```python
df
```

```python
inputs = df.drop('Salary', axis='columns')
```

```python
inputs
```

```python
targated = df['Salary']
```

```python
targated
```

```python
# importing sklearn lib

from sklearn.preprocessing import LabelEncoder
```

```python
lbl_enc_company = LabelEncoder()
lbl_enc_degignation = LabelEncoder()
lbl_enc_degree = LabelEncoder()
```

```python
inputs['Company_N'] = lbl_enc_company.fit_transform(inputs['Company'])
```

```python
inputs['Degignation_N'] = lbl_enc_degignation.fit_transform(inputs['Degignation'])
```

```python
inputs['Degree_N'] = lbl_enc_degree.fit_transform(inputs['Degree'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Company', 'Degignation', 'Degree'], axis='columns')
```

```python
inputs_n
```

```python
targated
```

```python
# importing decision tree classifier from sklearn

from sklearn import tree

dt_model = tree.DecisionTreeClassifier()
```

```python
dt_model.fit(inputs_n, targated)
```

```python
# check scores

dt_model.score(inputs_n, targated)
```

```python
# check whether salary >= 1 lac if doing job in google as computer engeener?
# Company | Post | Degree

dt_model.predict([[1, 1, 2]])
```
